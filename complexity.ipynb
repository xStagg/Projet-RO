```python
import numpy as np
import time
from nordouest import nord_ouest
from balashammer import balas_hammer
from marchepied import marche_pied_complet
import matplotlib.pyplot as plt
```

```python
def test_NO(n, cost, P, C):
    t0 = time.perf_counter()
    nord_ouest(n, n, cost, P, C)
    t1 = time.perf_counter()
    return t1-t0
```

```python
def test_BH(n, cost, P, C):
    t0 = time.perf_counter()
    balas_hammer(n, n, cost, P, C)
    t1 = time.perf_counter()
    return t1-t0
```

```python
def test_MPNO(n, cost, P, C):
    prop = nord_ouest(n, n, cost, P, C)
    t0 = time.perf_counter()
    marche_pied_complet(n, n, cost, P, C, prop, f2, f1)
    t1 = time.perf_counter()
    return t1-t0
```

```python
def test_MPBH(n, cost, P, C):
    prop = balas_hammer(n, n, cost, P, C)
    t0 = time.perf_counter()
    marche_pied_complet(n, n, cost, P, C, prop, f2, f1)
    t1 = time.perf_counter()
    return t1-t0
```

```python
def test_all(n):
    cost = np.random.randint(1, 101, (n, n), dtype=np.int32)
    temp = np.random.randint(1, 101, (n, n), dtype=np.int32)
    P, C = np.sum(temp, axis=1), np.sum(temp, axis=0)
    

    print(cost)
    print(P, C)

    tno = test_NO(n, cost, P, C)
    tbh = test_BH(n, cost, P, C)
    tmpno = test_MPNO(n, cost, P, C)
    tmpbh = test_MPBH(n, cost, P, C)
    return tno, tbh, tmpno, tmpbh
```

```python
def nuage_point(n):
    T_no = []
    T_bh = []
    T_mpno = []
    T_mpbh = []
    for _ in range(100):
        T = test_all(n)
        T_no.append(T[0])
        T_bh.append(T[1])
        T_mpno.append(T[2])
        T_mpbh.append(T[3])
    return T_no, T_bh, T_mpno, T_mpbh
```

```python
rng = np.random.default_rng(42)

def jitter_x(x, scale=0.04):
    # bruit multiplicatif adapté à une échelle log
    return x * np.exp(rng.uniform(-scale, scale, size=len(x)))
```

```python
# Les valeurs de n à tester (suivant l’énoncé)
N_values = [10]

all_no   = []
all_bh   = []
all_mpno = []
all_mpbh = []
all_n_no   = []
all_n_bh   = []
all_n_mpno = []
all_n_mpbh = []

for n in N_values:
    T_no, T_bh, T_mpno, T_mpbh = nuage_point(n)
    # répéter n 100 fois pour avoir les abscisses
    all_n_no.extend([n] * len(T_no))
    all_n_bh.extend([n] * len(T_bh))
    all_n_mpno.extend([n] * len(T_mpno))
    all_n_mpbh.extend([n] * len(T_mpbh))

    all_no.extend(T_no)
    all_bh.extend(T_bh)
    all_mpno.extend(T_mpno)
    all_mpbh.extend(T_mpbh)

all_n_no   = np.array(all_n_no)
all_n_bh   = np.array(all_n_bh)
all_n_mpno = np.array(all_n_mpno)
all_n_mpbh = np.array(all_n_mpbh)

all_no     = np.array(all_no)
all_bh     = np.array(all_bh)
all_mpno   = np.array(all_mpno)
all_mpbh   = np.array(all_mpbh)

all_no_mpno = all_no + all_mpno
all_bh_mpbh = all_bh + all_mpbh

print(all_no_mpno)
print(all_bh_mpbh)
```

```python
fig, ax = plt.subplots()

ax.scatter(jitter_x(all_n_no),   all_no,   c="r", s=10, alpha=0.6, label="t_NO(n)")
ax.scatter(jitter_x(all_n_bh),   all_bh,   c="g", s=10, alpha=0.6, label="t_BH(n)")
ax.scatter(jitter_x(all_n_mpno), all_mpno, c="w", s=10, alpha=0.6, label="t_MPNO(n)")
ax.scatter(jitter_x(all_n_mpbh), all_mpbh, c="k", s=10, alpha=0.6, label="t_MPBH(n)")
ax.scatter(jitter_x(all_n_no), all_no_mpno, c="m", s=10, label="t_NOMPNO(n)")
ax.scatter(jitter_x(all_n_bh), all_bh_mpbh, c="c", s=10, label="t_BHMPBH(n)")

ax.set_xscale("log")   # utile vu les valeurs de n (10 à 10^4)
ax.set_xlabel("n")
ax.set_ylabel("Temps d'exécution (s)")
ax.set_title("Nuages de points des temps mesurés")
ax.legend()

plt.show()
```

```python
def pire_cas(n):
    T_no, T_bh, T_mpno, T_mpbh = nuage_point(n)

    max_no = np.max(T_no)
    max_bh = np.max(T_bh)
    max_mpno = np.max(T_mpno)
    max_mpbh = np.max(T_mpbh)

    max_no_mpno = np.max(np.array(T_no) + np.array(T_mpno))
    max_bh_mpbh = np.max(np.array(T_bh) + np.array(T_mpbh))

    return max_no, max_bh, max_mpno, max_mpbh, max_no_mpno, max_bh_mpbh
```

```python
N_values = [10, 11, 12, 13, 14, 15]

W_no = []
W_bh = []
W_mpno = []
W_mpbh = []
W_no_mpno = []
W_bh_mpbh = []

for n in N_values:
    a, b, c, d, e, f = pire_cas(n)
    W_no.append(a)
    W_bh.append(b)
    W_mpno.append(c)
    W_mpbh.append(d)
    W_no_mpno.append(e)
    W_bh_mpbh.append(f)
```

```python
fig, ax = plt.subplots()

ax.plot(N_values, W_no, "o-", label="max θ_NO(n)", color="r")
ax.plot(N_values, W_bh, "o-", label="max θ_BH(n)", color="g")
ax.plot(N_values, W_mpno, "o-", label="max t_NO(n)", color="b")
ax.plot(N_values, W_mpbh, "o-", label="max t_BH(n)", color="k")
ax.plot(N_values, W_no_mpno, "o-", label="max (θ_NO+t_NO)(n)", color="m")
ax.plot(N_values, W_bh_mpbh, "o-", label="max (θ_BH+t_BH)(n)", color="c")

ax.set_xscale("log")
ax.set_yscale("log")   # très utile pour reconnaître la complexité
ax.set_xlabel("n")
ax.set_ylabel("Pire temps mesuré")
ax.set_title("Complexité expérimentale dans le pire des cas")
ax.legend()

plt.show()
```

```python
def comparaison_pire_cas(n):
    T_no, T_bh, T_mpno, T_mpbh = nuage_point(n)

    T_no = np.array(T_no)
    T_bh = np.array(T_bh)
    T_mpno = np.array(T_mpno)
    T_mpbh = np.array(T_mpbh)

    ratio = (T_no + T_mpno) / (T_bh + T_mpbh)
    return np.max(ratio)
```

```python
N_values = [10]
R_max = []

for n in N_values:
    r = comparaison_pire_cas(n)
    R_max.append(r)
```

```python
fig, ax = plt.subplots()

ax.plot(N_values, R_max, "o-", color="purple")
ax.set_xscale("log")
ax.set_xlabel("n")
ax.set_ylabel(r"$\max \frac{t_{NO}(n)+\theta_{NO}(n)}{t_{BH}(n)+\theta_{BH}(n)}$")
ax.set_title("Comparaison du pire cas")
plt.show()
```
